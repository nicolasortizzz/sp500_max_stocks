**Importancia de la Correlaci√≥n en Inversiones:**  
La correlaci√≥n es una medida estad√≠stica que indica c√≥mo se mueven juntas dos variables. El coeficiente de correlaci√≥n, \( r \), se calcula con la siguiente f√≥rmula:

$$ r = \frac{\sum (x_i - \overline{x})(y_i - \overline{y})}{\sqrt{\sum (x_i - \overline{x})^2 \sum (y_i - \overline{y})^2}} $$

Donde $$x_i $$ y $$ y_i $$ son los valores individuales de las variables y $$ \overline{x} $$ y $$ \overline{y} $$ son las medias de esas variables. Un valor de 1 indica una correlaci√≥n positiva perfecta, mientras que un valor de -1 indica una correlaci√≥n negativa perfecta. Un valor de 0 sugiere que no hay correlaci√≥n lineal entre las variables.





```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

etfs = [
    'XLE', 'XLF', 'XLU', 'XLI', 'GDX', 'XLK', 'XLV', 'XLY', 'XLP', 'XLB', 
    'XOP', 'IYR', 'XHB', 'ITB', 'VNQ', 'GDXJ', 'IYE', 'OIH', 'XME', 'XRT', 
    'SMH', 'IBB', 'KBE', 'KRE', 'XTL'
]

start_date = '2010-01-01'
end_date = '2024-12-31'


data = yf.download(etfs, start=start_date, end=end_date)['Adj Close']


returns = data.pct_change().dropna()  #NAN #Null


correlation_returns = returns.corr()

#Prepararnos para graficar

plt.figure(figsize=(14, 12))


sns.heatmap(correlation_returns, annot=True, fmt=".2f", cmap='coolwarm')


plt.title('Correlaci√≥n de los retornos diarios de los ETFs seleccionados desde 2010')


plt.xticks(rotation=45)
plt.yticks(rotation=45)


plt.tight_layout()


plt.show()

```
